```python
'''Q2 What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.'''
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Display the plot
plt.show()
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])'''
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # First subplot
plt.plot(x, y1)
plt.title('Line 1: y = 0, 100, 200, 300, 400, 500')

plt.subplot(2, 2, 2)  # Second subplot
plt.plot(x, y2)
plt.title('Line 2: y = 50, 20, 40, 20, 60, 70')

plt.subplot(2, 2, 3)  # Third subplot
plt.plot(x, y3)
plt.title('Line 3: y = 10, 20, 30, 40, 50, 60')

plt.subplot(2, 2, 4)  # Fourth subplot
plt.plot(x, y4)
plt.title('Line 4: y = 200, 350, 250, 550, 450, 150')

# Display the plots
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
A bar plot (also known as a bar chart) is a type of graph that represents categorical data with rectangular bars. The length of each bar is proportional to the value it represents. Bar plots can be vertical (with bars extending upward) or horizontal (with bars extending to the right or left).

Why is it used?
Bar plots are useful for comparing discrete categories, showing the frequency or magnitude of different groups.
They are easy to read and interpret, making them ideal for comparing quantities across categories.
It is commonly used in business, finance, and other domains to visualize data like sales, profits, or quantities across different entities or groups.'''
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit)
plt.title('Company Profits (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit in USD')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title('Company Profits (Horizontal Bar Plot)')
plt.xlabel('Profit in USD')
plt.ylabel('Company')
plt.show()

```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
A box plot (also known as a box-and-whisker plot) is a graphical representation of a dataset's distribution through its quartiles. It highlights the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum of a dataset, along with any potential outliers.

Why is it used?
Visualize the distribution: A box plot provides a simple way to visualize the spread and skewness of a dataset.
Identify outliers: It highlights outliers (data points significantly higher or lower than the rest of the data).
Compare datasets: Multiple box plots can be compared side by side to analyze differences in distributions across multiple groups.'''
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)  # mean = 100, std = 10, sample size = 200
box2 = np.random.normal(90, 20, 200)   # mean = 90, std = 20, sample size = 200

# Plotting the box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Adding title and labels
plt.title('Box Plot of Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Values')

# Display the plot
plt.show()
```
